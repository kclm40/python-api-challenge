### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
#configure gmaps
gmaps.configure(api_key=g_key)

#use the Lat and Lng as locations and Humidity as the weight
locations = file[["Latitude", "Longitude"]]
weight = file["Humidity"]

```

```python
#Add Heatmap layer to map
fig = gmaps.figure(center = [0,0], zoom_level =2)

heat_layer = gmaps.heatmap_layer(locations, weight, dissipating=False, max_intensity =100, point_radius = 4)

fig.add_layer(heat_layer)
fig
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
hotel_df = limit_file
hotel_df.reset_index(drop=True)
```

```python
#add hotel name to the Dataframe
hotel_df["Hotel Name"] = ""
```

```python
#set parameters
params= {"radius": 5000,
         "keyword": "hotels",
         "types": "hotel",
         "key": g_key   
        }
```

```python
#use iterrrows to call to the google places API
for index, row in hotel_df.iterrows():
    
    latitude = row["Latitude"]
    longitude = row["Longitude"]
    
    #update location each loop
    
    params['location'] = f'{latitude},{longitude}'
    
    #create base url
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    
    #request
    hotel_location = requests.get(base_url, params=params)
    
    #json request
    hotel_location = hotel_location.json()
    
    
    try:
        
        hotel_df.loc[index, "Hotel Name"] = hotel_location["results"][0]["name"]
    
    except:
        print('Missing')
        hotel_df.loc[index, "Hotel Name"] = "NA"

```
